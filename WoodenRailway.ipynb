```python
class Trenecito(Problem):
    
    def __init__(self):
        self.initial_state = State.value_of(INITIAL_STATE)
        self.goal_state = State.value_of(GOAL_STATE)
        super().__init__(self.initial_state, self.goal_state) 
        
    def actions(self, state):
        return ['ponerCI','ponerR','ponerCD','ponerJoin','ponerFork','ponerForkR','ponerForkCI','ponerForkCD']    

    def result(self, state, action):
        coordx = state.coordx
        coordy = state.coordy
        fx = state.fx
        fy = state.fy
        jx = state.jx
        jy = state.jy
        rectas = state.rectas
        curvas = state.curvas
        forks = state.forks
        joins = state.joins
        orientacion = state.orientacion
        seq = state.seq
        next_state =  State(coordx,coordy,fx,fy,jx,jy,rectas,curvas,forks,joins,orientacion,seq)
        
        if state.orientacion == "N":  #NORTE
            if action == "ponerR" and state.rectas!=0 :  
              #  seq.append("R")
                next_state.coordy = next_state.coordy + 1 
                next_state.rectas = next_state.rectas - 1            
            
            elif action == "ponerCI" and state.curvas!=0 :
              #  next_state.seq.append("CI")
                next_state.coordx = next_state.coordx - 1
                next_state.coordy = next_state.coordy + 1
                next_state.curvas = next_state.curvas - 1
                next_state.orientacion = "NO"
                
            elif action == "ponerCD"and state.curvas!=0 :
              #  next_state.seq.append("CD")
                next_state.coordx = next_state.coordx + 1
                next_state.coordy = next_state.coordy + 1
                next_state.curvas = next_state.curvas - 1
                next_state.orientacion = "NE"
                
                
                
        elif state.orientacion == "S" :   #SUR
            if action == "ponerR" and state.rectas!=0 :  
             #   seq.append("R")
                next_state.coordy = next_state.coordy - 1 
                next_state.rectas = next_state.rectas - 1
                    
            elif action == "ponerCI" and state.curvas!=0:
             #   next_state.seq.append("CI")
                next_state.coordx = next_state.coordx + 1 
                next_state.coordy = next_state.coordy - 1
                next_state.curvas = next_state.curvas - 1
                next_state.orientacion = "SE"
                
            elif action == "ponerCD" and state.curvas!=0:
              #  next_state.seq.append("CD")
                next_state.coordx = next_state.coordx - 1  
                next_state.coordy = next_state.coordy - 1
                next_state.curvas = next_state.curvas - 1
                next_state.orientacion = "SO"
                
                

                
        elif state.orientacion == "E": #ESTE
            if action == "ponerR" and state.rectas!=0:  
             #   seq.append("R")
                next_state.coordx = next_state.coordx + 1  
                next_state.rectas = next_state.rectas - 1
                
            elif action == "ponerCI" and state.curvas!=0:
              #  next_state.seq.append("CI")
                next_state.coordx = next_state.coordx + 1
                next_state.coordy = next_state.coordy + 1
                next_state.curvas = next_state.curvas - 1
                next_state.orientacion = "NE"
                
            elif action == "ponerCD" and state.curvas!=0:
             #   next_state.seq.append("CD")
                next_state.coordx = next_state.coordx + 1
                next_state.coordy = next_state.coordy - 1
                next_state.curvas = next_state.curvas - 1
                next_state.orientacion = "SE"
                
            elif action == "ponerFork" and state.forks!=0:
                next_state.fx = next_state.coordx + 1
                next_state.coordx = next_state.coordx + 1
                next_state.fy = next_state.coordy - 1
                next_state.forks = next_state.forks - 1
                
            elif action == "ponerForkR" and state.rectas!=0 and state.forks==0 and state.joins==1:
                next_state.fx = next_state.fx + 1
                next_state.rectas = next_state.rectas - 1
                
            elif action == "ponerForkCI" and state.curvas!=0 and state.forks==0 and state.joins==1:
                next_state.fx = next_state.fx + 1
                next_state.fy = next_state.fy - 1
                next_state.curvas = next_state.curvas - 1

            elif action == "ponerForkCD" and state.curvas!=0 and state.forks==0 and state.joins==1:  
                next_state.fx = next_state.fx + 1
                next_state.fy = next_state.fy + 1
                next_state.curvas = next_state.curvas - 1

            elif action == "ponerJoin" and state.joins!=0 and state.forks == 0 :
                next_state.jx = next_state.coordx 
                next_state.coordx = next_state.coordx + 1
                next_state.jy = next_state.coordy - 1
                next_state.joins = next_state.joins - 1
                
                
        elif state.orientacion == "O":  #OESTE
            
            if action == "ponerR" and state.rectas!=0:  
              #  seq.append("R")
                next_state.coordx = next_state.coordx - 1  
                next_state.rectas = next_state.rectas - 1  
                
            elif action == "ponerCI" and state.curvas!=0:
             #   next_state.seq.append("CI")
                next_state.coordx = next_state.coordx - 1
                next_state.coordy = next_state.coordy - 1
                next_state.curvas = next_state.curvas - 1
                next_state.orientacion = "SO"
                
            elif action == "ponerCD" and state.curvas!=0:
             #   next_state.seq.append("CD")
                next_state.coordx = next_state.coordx - 1
                next_state.coordy = next_state.coordy + 1
                next_state.curvas = next_state.curvas - 1
                next_state.orientacion = "NO"
                
                
                
        elif state.orientacion == "NE":  #NORESTE    
            
            if action == "ponerR" and state.rectas!=0:  
             #   seq.append("R")
                next_state.coordx = next_state.coordx + 1  
                next_state.coordy = next_state.coordy + 1
                next_state.rectas = next_state.rectas - 1       
                
            elif action == "ponerCI" and state.curvas!=0 :
             #   next_state.seq.append("CI")
                next_state.coordx = next_state.coordx + 1
                next_state.coordy = next_state.coordy + 1
                next_state.curvas = next_state.curvas - 1
                next_state.orientacion = "N"
                
            elif action == "ponerCD" and state.curvas!=0:
             #   next_state.seq.append("CD")
                next_state.coordx = next_state.coordx + 1
                next_state.coordy = next_state.coordy - 1
                next_state.curvas = next_state.curvas - 1
                next_state.orientacion = "E"
                            
                
        elif state.orientacion == "NO":  #NOROESTE   
            
            if action == "ponerR" and state.rectas!=0:  
               # seq.append("R")
                next_state.coordx = next_state.coordx - 1  
                next_state.coordy = next_state.coordy + 1
                next_state.rectas = next_state.rectas - 1
                
            elif action == "ponerCI" and state.curvas!=0:
               # next_state.seq.append("CI")
                next_state.coordx = next_state.coordx - 1
                next_state.coordy = next_state.coordy + 1
                next_state.curvas = next_state.curvas - 1
                next_state.orientacion = "O"
                
            elif action == "ponerCD" and state.curvas!=0:
                #next_state.seq.append("CD")
                next_state.coordx = next_state.coordx + 1
                next_state.coordy = next_state.coordy - 1
                next_state.curvas = next_state.curvas - 1
                next_state.orientacion = "E"

                
        elif state.orientacion == "SE":  #SURESTE
            
            if action == "ponerR" and state.rectas!=0:  
                #seq.append("R")
                next_state.coordx = next_state.coordx + 1  
                next_state.coordy = next_state.coordy - 1
                next_state.rectas = next_state.rectas - 1
                
            elif action == "ponerCI" and state.curvas!=0:
               # next_state.seq.append("CI")
                next_state.coordx = next_state.coordx + 1
                next_state.coordy = next_state.coordy - 1
                next_state.curvas = next_state.curvas - 1
                next_state.orientacion = "E"
                
            elif action == "ponerCD" and state.curvas!=0:
              #  next_state.seq.append("CD")
                next_state.coordx = next_state.coordx + 1
                next_state.coordy = next_state.coordy - 1
                next_state.curvas = next_state.curvas - 1
                next_state.orientacion = "S"  
                
                            
                
        elif state.orientacion == "SO":  #SUROESTE
            
            if action == "ponerR" and state.rectas!=0:  
             #  seq.append("R")
                next_state.coordx = next_state.coordx - 1  
                next_state.coordy = next_state.coordy - 1
                next_state.rectas = next_state.rectas - 1
                
            elif action == "ponerCI" and state.curvas!=0 :
               
            #next_state.seq.append("CI")
                next_state.coordx = next_state.coordx - 1
                next_state.coordy = next_state.coordy - 1
                next_state.curvas = next_state.curvas - 1
                next_state.orientacion = "S"
                
            elif action == "ponerCD" and state.curvas!=0:
               # next_state.seq.append("CD")
                next_state.coordx = next_state.coordx - 1
                next_state.coordy = next_state.coordy - 1
                next_state.curvas = next_state.curvas - 1
                next_state.orientacion = "O"
               
                    
        return next_state
                
    def goal_test(self, state):
        if isinstance(self.goal, list):
            return state in self.goal  
        else:
            return state == self.goal  
    def path_cost(self, cost, state1, action, state2):
        return cost+1
    
```

```python
class State:
    orientacion = ""
    def __init__(self,coordx,coordy,fx,fy,jx,jy,rectas,curvas,forks,joins,orientacion,seq):
        self.value = (coordx,coordy,fx,fy,jx,jy,rectas,curvas,forks,joins,orientacion,seq)
        self.coordx = coordx
        self.coordy = coordy
        self.fx = fx
        self.fy = fy
        self.jx = jx
        self.jy = jy
        self.rectas = rectas
        self.curvas = curvas
        self.forks = forks
        self.joins = joins
        self.orientacion = orientacion
        self.seq = seq
        
    @classmethod
    def value_of(cls, state):
        if not cls.__is_valid_tuple(state):
            raise ValueError(str(state) + " must be a tuple with 12 elements.")
        return State(state[0], state[1], state[2], state[3], state[4], state[5],state[6],state[7],state[8],state[9],state[10],state[11])

    def is_valid(self):
        if contains_negative(self.value):
            return False
        else:
            return True
    
    @staticmethod
    def __is_valid_tuple(other):
        return isinstance(other, tuple) and len(other) == 12

    @staticmethod
    def __get_invalid_operand_error(other):
        return str(other) + " must be an instance of State or a tuple of length 12."

    def __repr__(self):
        return '<State {}>'.format(self.value)

    def __str__(self):
        return '<State {}>'.format(self.value)

    def __lt__(self, other):
        self.__ensure_instance_of_state(other)
        return self.value < other.value

    def __eq__(self, other):
        return isinstance(other, State) and self.value == other.value

    def __hash__(self):
        return hash(self.value)

    @staticmethod
    def __ensure_instance_of_state(other):
        if not isinstance(other, State):
            raise ValueError(str(other) + " must be an instance of State")
    
```
